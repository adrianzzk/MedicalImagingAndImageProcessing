## 二值化

```python
import SimpleITK as sitk
from Niiplot import multi_slice_viewer
```

```python
img=sitk.ReadImage('../data/lung.nii.gz')
```

#### 这里拿骨头作为实例 （ct值为500-1000）

```python
seg_img=sitk.BinaryThreshold(img,500,1000)     
```

#### 下面为平化处理

```python
seg_img=sitk.BinaryMorphologicalClosing(seg_img)
seg_img=sitk.BinaryMorphologicalClosing(seg_img)
seg_img=sitk.BinaryMorphologicalClosing(seg_img)
```

```python
multi_slice_viewer(img)
```

```python
sitk.WriteImage(seg_img,'../result/bone.nii.gz')
```

## 区域生长

```python
import SimpleITK as sitk
from Niiplot import multi_slice_viewer
```

```python
img=sitk.ReadImage("../data/lung.nii.gz")
```

#### 生成一个种子（多个也可以，只要是在对应的区域内就可以）

```python
slt=[]
slt.append([332,295,52])
slt.append([121,326,37])
```

```python
img=sitk.ConnectedThreshold(img,slt,-900,-400)
img=sitk.BinaryMorphologicalClosing(img)
```

```python
multi_slice_viewer(img)
```

```python
sitk.WriteImage(img,'../result/region_growth.nii.gz')
```
