# 滤波

### 普通的滤波

```python
import SimpleITK as sitk
from Niiplot import multi_slice_viewer
```

```python
img=sitk.ReadImage("../data/lung.nii.gz")
img=sitk.Cast(img,sitk.sitkFloat32)
img=sitk.SobelEdgeDetection(img)
```

```python
multi_slice_viewer(img)
```

```python
sitk.WriteImage(img,'../result/lung_filter.nii.gz')
```

### gaussian滤波

```python
import SimpleITK as sitk
from Niiplot import multi_slice_viewer
```

```python
img=sitk.ReadImage('../data/lung.nii.gz')
img=sitk.Cast(img,sitk.sitkFloat32)
img=sitk.DiscreteGaussian(img,[3,3,3])
```

```python
multi_slice_viewer(img)
```

```python
sitk.WriteImage(img, '../result/lung_gaussain.nii.gz')
```

### 将上述两种操作连续进行

```python
import SimpleITK as sitk
from Niiplot import multi_slice_viewer
```

```python
img=sitk.ReadImage('../data/lung.nii.gz')

img=sitk.Cast(img,sitk.sitkFloat32)
img=sitk.DiscreteGaussian(img,[3,3,3])
img=sitk.SobelEdgeDetection(img)
```

```python
multi_slice_viewer(img)
```

```python
sitk.WriteImage(img,'../result/lung_edege.nii.gz')
```
